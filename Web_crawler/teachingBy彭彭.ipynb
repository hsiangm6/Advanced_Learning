```python
#Web crawler 基本
#抓取ppt電影版的html原始碼
# https://www.youtube.com/watch?v=9Z9xKWfNo7k&list=RDCMUCguZS-y7codLSt6vpkVdnKg&index=1
import urllib.request as req

url="https://www.ptt.cc/bbs/movie/index.html"
#建立request物件，再利用物件打開網址，傅家request headers的資訊
request=req.Request(url, headers={ 
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"
})
with req.urlopen(request) as response:
    data=response.read().decode("utf-8")


# 解析原始碼，取得每篇文章的標題
import bs4
root=bs4.BeautifulSoup(data, "html.parser") #讓beautifulsoup4協助我們解析html格式
print(root.title)
print(root.title.string)
post_title=root.find("div", class_="title") #find會幫助尋找其中一個class="title"的div標籤
print(post_title)
print(post_title.a.string, "\n")
post_titles=root.find_all("div", class_="title") #find會幫助尋找其中一個class="title"的div標籤，如果條件是id則id="title"
for title in post_titles:
    if title.a != None: #如果標題包含a標籤(沒有被刪除)
        print(title.a.string)

```

```python
#web crawler- cookie(連線時放在request headers中送出)
#追蹤下一頁網頁連結，連續抓取頁面
#抓取ppt八卦版的html
# https://www.youtube.com/watch?v=BEA7F9ExiPY&list=RDCMUCguZS-y7codLSt6vpkVdnKg&index=2
import urllib.request as req

url="https://www.ptt.cc/bbs/Gossiping/index.html"
def getData(url):
    #建立request物件，再利用物件打開網址，傅家request headers的資訊
    request=req.Request(url, headers={ 
        #cookie在network的其中一檔案中的request header
        "cookie":"over18=1", #加入cookie告訴網頁我有over18
        #user-agent在network的其中一檔案中的request header
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"
    })
    with req.urlopen(request) as response:
        data=response.read().decode("utf-8")


    # 解析原始碼，取得每篇文章的標題
    import bs4
    root=bs4.BeautifulSoup(data, "html.parser") #讓beautifulsoup4協助我們解析html
    post_titles=root.find_all("div", class_="title") #find會幫助尋找其中一個class="title"的div標籤，如果條件是id則id="title"
    for title in post_titles:
        if title.a != None: #如果標題包含a標籤(沒有被刪除)
            print(title.a.string)

    #抓取上一頁的網址
    nextLink=root.find("a", string="‹ 上頁") #找到內文是‹ 上頁 的a標籤
    return nextLink["href"] #中括號內是屬性名稱

#抓取一個頁面的標題
pageURL="https://www.ptt.cc/bbs/Gossiping/index.html"
count=0
while count<3: #抓取3頁的標題
    pageURL="https://www.ptt.cc"+getData(pageURL)
    count+=1
```
